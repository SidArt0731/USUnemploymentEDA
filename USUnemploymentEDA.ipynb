# Dataset of Unemployment rate from 1948-2021.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

### Loading Dataset

```python
unemp = pd.read_csv('data/unemployment_rate_data.csv')

unemp.head()
```

## As seen in the data there's Overall Unemployment Rate, Gender specific Unemployment Rate, Unemployment Rate according to age groups

### As per research, analyzing for three specific periods:

### 1980-1990's economic recession that affected the world mostly beacuse of 1979 energy crises caused Iranian Revolution which caused disruption in global oil supply

### 2010's Great Recession specifically during period from 2007-2009 (economic downturn) after boating of us housing bubble and the global financial crises

### Post Covid Data i.e. 2020's economic crises of Corona Virus

```python
unemp.info()
```

### Converting the date col into date-time format

```python
unemp['date'] = pd.to_datetime(unemp['date'])
unemp.info()
```

```python
unemp.describe()
```

### As seen above there are more unemployed people between the age of 16-17

```python
unemp.describe(percentiles=np.arange(0, 1.0, 0.1))
```

### As seen in the above indepth analysis Women are more umemployed than Men

```python
plt.figure(figsize=(15,5))
plt.plot(unemp['date'],unemp['unrate'])
```

### As seen above unemployment peaked in 2020-21, as second instance ,as obvious, is 2010s and third is 1980s

### Overall Unemployment rate for women

```python
plt.figure(figsize=(15,5))
sns.lineplot(data=unemp,x='date',y='unrate_women',color='red')
```

### Overall Unemployment rate for men

```python
plt.figure(figsize=(15,5))
sns.lineplot(data=unemp,x='date',y='unrate_men',color='blue')
```

## Unemployment rates in 1980-1990

```python
oil_disruption80s = unemp[(unemp['date']>="1/1/1980")&(unemp["date"]<='1/1/1984')]
plt.figure(figsize=(15,5))
sns.lineplot(data=oil_disruption80s,x='date',y='unrate')
```

### We can see unemployment rate peaked from 1982-1983 specifically

### Unemployment rate for Women between Jan 1982 to Dec 1983

```python
plt.figure(figsize=(15,5))
sns.lineplot(data=oil_disruption80s,x='date',y='unrate_women',color='red')
```

```python
oil_disruption80s['unrate_women'].describe()
```

### As seen above the unemployment rate for women started increasing from Jul 1982 and peaked during Jan 1983 and then gradually decreased till Apr and once again peaked in Jul

### Unemployment rate for Men between Jan 1982 to Dec 1983

```python
plt.figure(figsize=(15,5))
sns.lineplot(data=oil_disruption80s,x='date',y='unrate_men',color='blue')
```

```python
oil_disruption80s['unrate_men'].describe()
```

### As seen above for men, unemployment rate started increasing from aroung Jul/Aug 1982 and peaked during Jan-Apr 1983

### Segregating data into different age groups

```python
plt.figure(figsize=(15,10))
sns.lineplot(data=oil_disruption80s,x='date',y='unrate_16_to_17',label='unrate_16_to_17')
sns.lineplot(data=oil_disruption80s,x='date',y='unrate_18_to_19',label='unrate_18_to_19')
sns.lineplot(data=oil_disruption80s,x='date',y='unrate_20_to_24',label='unrate_20_to_24')
sns.lineplot(data=oil_disruption80s,x='date',y='unrate_25_to_34',label='unrate_25_to_34')
sns.lineplot(data=oil_disruption80s,x='date',y='unrate_35_to_44',label='unrate_35_to_44')
sns.lineplot(data=oil_disruption80s,x='date',y='unrate_45_to_54',label='unrate_45_to_54')
sns.lineplot(data=oil_disruption80s,x='date',y='unrate_55_over',label='unrate_55_over')
```

```python
oil_disruption80s[['unrate_16_to_17','unrate_18_to_19', 'unrate_20_to_24', 'unrate_25_to_34','unrate_35_to_44', 'unrate_45_to_54', 'unrate_55_over']].describe()
```

### Once again we can see people aged around 16-19 were much more unemployed during 1980s

## 2010's recession analysis

```python
recession = unemp[(unemp['date']>="01-01-2009") & (unemp["date"]<="12-30-2014")]
recession.tail()
```

### Overall recession in 2010s

```python
plt.figure(figsize=(15,5))
sns.lineplot(data=recession,x='date',y='unrate')
```

### Rescession in 2010s on men

```python
plt.figure(figsize=(15,5))
sns.lineplot(data=recession,x='date',y='unrate_men',label='men',color='blue')
plt.legend
```

```python
recession['unrate_men'].describe()
```

### as seen above recession on men during 2010s starts during 2009 and peaked during 2010 then gradually starts decreasing

### Rescession in 2010s on women

```python
plt.figure(figsize=(15,5))
sns.lineplot(data=recession,x='date',y='unrate_women',label='women',color='red')
plt.legend
```

```python
recession['unrate_women'].describe()
```

### As seen above recession on women is on the peak during the duration of 2009-2013 

### 2010s Recession on age groups

```python
plt.figure(figsize=(15,10))
sns.lineplot(data=recession,x='date',y='unrate_16_to_17',label='unrate_16_to_17')
sns.lineplot(data=recession,x='date',y='unrate_18_to_19',label='unrate_18_to_19')
sns.lineplot(data=recession,x='date',y='unrate_20_to_24',label='unrate_20_to_24')
sns.lineplot(data=recession,x='date',y='unrate_25_to_34',label='unrate_25_to_34')
sns.lineplot(data=recession,x='date',y='unrate_35_to_44',label='unrate_35_to_44')
sns.lineplot(data=recession,x='date',y='unrate_45_to_54',label='unrate_45_to_54')
sns.lineplot(data=recession,x='date',y='unrate_55_over',label='unrate_55_over')
```

```python
recession[['unrate_16_to_17','unrate_18_to_19', 'unrate_20_to_24', 'unrate_25_to_34','unrate_35_to_44', 'unrate_45_to_54', 'unrate_55_over']].describe()
```

### Once again we can see people aged around 16-19 were much more unemployed during 2010s

### 2020s Economic crises due to Corona

```python
covid_19 = unemp[(unemp['date']>="01-01-2020")]
covid_19.head()
```

### Unemployment rate for men

```python
plt.figure(figsize=(15,5))
sns.lineplot(data=covid_19,x='date',y='unrate_men',label='men',color='blue')
plt.legend
```

### Around March 2020 unemployment peaked for men at around 13% and as seen it gradually started decreasing from July 2021

```python
covid_19['unrate_men'].describe()
```

### Around March 2020 unemployment peaked for men at around 13% and as seen it gradually started decreasing from July 2021

### Unemployment rate for men

```python
plt.figure(figsize=(15,5))
sns.lineplot(data=covid_19,x='date',y='unrate_women',label='women',color='blue')
plt.legend
```

### Around March 2020 unemployment peaked for women as well but at around 15% and as seen it gradually started decreasing from July 2021

### Segregating data into different age groups

```python
plt.figure(figsize=(15,10))
sns.lineplot(data=covid_19,x='date',y='unrate_16_to_17',label='unrate_16_to_17')
sns.lineplot(data=covid_19,x='date',y='unrate_18_to_19',label='unrate_18_to_19')
sns.lineplot(data=covid_19,x='date',y='unrate_20_to_24',label='unrate_20_to_24')
sns.lineplot(data=covid_19,x='date',y='unrate_25_to_34',label='unrate_25_to_34')
sns.lineplot(data=covid_19,x='date',y='unrate_35_to_44',label='unrate_35_to_44')
sns.lineplot(data=covid_19,x='date',y='unrate_45_to_54',label='unrate_45_to_54')
sns.lineplot(data=covid_19,x='date',y='unrate_55_over',label='unrate_55_over')
```
